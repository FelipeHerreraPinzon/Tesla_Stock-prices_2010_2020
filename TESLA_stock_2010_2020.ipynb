```python

import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos desde tu base de datos
url = 'https://raw.githubusercontent.com/FelipeHerreraPinzon/Tesla_Stock-prices_2010_2020/main/TSLA.csv'
df = pd.read_csv(url)
df
```

```python
# Convertir la columna Date al formato de fecha
df['Date'] = pd.to_datetime(df['Date'])

# Crear columnas separadas para el mes y el año
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month

# Crear una nueva columna para mm/yyyy
df['Month_Year'] = df['Date'].dt.to_period('M')

# Calcular el promedio de cada mes
monthly_avg = df.groupby('Month_Year').mean()

# Reordenar las columnas
monthly_avg = monthly_avg[['Open', 'Close', 'Low', 'High', 'Adj Close', 'Volume']]

# Mostrar el DataFrame con los promedios mensuales
monthly_avg

#exportarlo a CSV
#monthly_avg.to_csv('TESLA_2010_2020.csv')
```
